```python
import nltk

def ngram_distance(s1, s2, n):
    # Tokenizar las cadenas de texto en N-gramas
    s1_ngrams = set(nltk.ngrams(s1, n))
    s2_ngrams = set(nltk.ngrams(s2, n))
    
    # Calcular la similitud entre los N-gramas
    common = len(s1_ngrams & s2_ngrams)
    total = len(s1_ngrams) + len(s2_ngrams)
    return 1 - (2 * common) / (total - common)

string1 = "casa"
string2 = "calle"
distance = ngram_distance(string1, string2, 2)
print (distance)  # Output: 0.6666666666666667
```
