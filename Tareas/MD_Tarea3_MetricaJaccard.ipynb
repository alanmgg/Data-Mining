```python
def jaccard_distance(s1, s2):
    # Convertir las cadenas en conjuntos de palabras
    s1_jaccard = set(s1.split())
    s2_jaccard = set(s2.split())
    
    # Calcular la distancia de Jaccard
    intersection = len(s1_jaccard.intersection(s2_jaccard))
    union = len(s1_jaccard.union(s2_jaccard))
    distance = 1 - (intersection / union)
    return distance

string1 = "casa bonita"
string2 = "casa grande"
distance = jaccard_distance(string1, string2)
print(distance)  # Output: 0.6666666666666667
```
