```python
def levenshtein_distance(s1, s2):
    s1_levenshtein = len(s1)
    s2_levenshtein = len(s2)
    distance = [[0] * (s2_levenshtein + 1) for _ in range(s1_levenshtein + 1)]
    for i in range(s1_levenshtein + 1):
        distance[i][0] = i
    for j in range(s2_levenshtein + 1):
        distance[0][j] = j
    for j in range(1, s2_levenshtein + 1):
        for i in range(1, s1_levenshtein + 1):
            if s1[i - 1] == s2[j - 1]:
                distance[i][j] = distance[i - 1][j - 1]
            else:
                distance[i][j] = min(distance[i - 1][j], distance[i][j - 1], distance[i - 1][j - 1]) + 1
    return distance[s1_levenshtein][s2_levenshtein]
  
string1 = "casa"
string2 = "calle"
distance = levenshtein_distance(string1, string2)
print (distance)  # Output: 3
```
