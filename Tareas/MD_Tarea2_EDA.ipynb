```python
import pandas as pd                     # Para la manipulación y análisis de datos
import numpy as np                      # Para crear vectores y matrices n dimensionales
import matplotlib.pyplot as plt         # Para la generación de gráficas a partir de los datos
import seaborn as sns                   # Para la visualización de datos basado en matplotlib
%matplotlib inline
# Para generar imagenes dentro del cuaderno
```

```python
url = "https://raw.githubusercontent.com/alanmgg/Data-Mining/main/Datos/courses.csv"
data_courses = pd.read_csv(url)
data_courses
```

```python
data_courses.hist(figsize=(14,14), xrot=45)
plt.show()
```

```python
data_courses.describe()
```

```python
variable = ['rating', 'num_reviews', 'num_published_lectures']
for col in variable:
  sns.boxplot(x=col, data=data_courses)
  plt.show()
```

```python
data_courses.describe(include='object')
```

```python
for col in data_courses.select_dtypes(include='object'):
  if data_courses[col].nunique()<10:sns.countplot(y=col, data=data_courses)
  plt.show()
```

```python
for col in data_courses.select_dtypes(include='object'):
  if data_courses[col].nunique() < 10:
    display(data_courses.groupby(col).agg(['mean']))
```

```python
data_courses.corr()
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(data_courses.corr(), cmap='RdBu_r', annot=True)
plt.show()
```

```python
plt.figure(figsize=(14,7))
MatrizInf = np.triu(data_courses.corr())
sns.heatmap(data_courses.corr(), cmap='RdBu_r', annot=True, mask=MatrizInf)
plt.show()
```

```python
data_nan = data_courses.dropna()
data_nan
```

```python
data_sesgo = data_nan[data_nan['rating'] > 3] 
data_sesgo = data_sesgo[data_sesgo['num_reviews'] < 250]
data_sesgo = data_sesgo[data_sesgo['num_published_lectures'] < 100]
```

```python
data_sesgo.hist(figsize=(14,14), xrot=45)
plt.show()
```
